#Annotating Images
In this notebook we will cover how to annote images using OpenCV




*   Draw line
*   Draw Circle

#Draw Line
img=cv2.line(img,pt1,pt2,color[,thickness[, linetypw[,shift]]]])


```python
#Line
image_Line=img.copy()
#Draw line
cv2.line(image_Line,(200,100),(400,500),(255,0,255),thickness=50,lineType=cv2.LINE_AA);
plt.imshow(image_Line[:,:,::-1])
```

#Draw Circle
img=cv2.circle(img,center,radius,color[.....]


```python
#Circle
img_circle=img[:,:,::-1].copy()
cv2.circle(img_circle,(1500,1000),400,(0,0,255),thickness=20)
plt.imshow(img_circle)
plt.title("Circle")
```

#Draw Rectangle
img=cv2.rectangle=(img,pt1,pt2,color....)


```python
#Rectangle
img_rectangle=img[:,:,::-1].copy()
cv2.rectangle(img_rectangle,(1000,1700),(400,500),(255,0,255),thickness=50,lineType=cv2.LINE_AA);
plt.imshow(img_rectangle)
```
